```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

```

```python
# Reading the data and printing the first 5 registers
data = pd.read_csv('insurance.csv')
data.head()
```

Plot a scatter plot with age on the x-axis and charges on the y-axis.

```python
# Scatter plot Age vs Charges
x = data[['age']].values
y = data[['charges']].values
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5, edgecolor='k', label='Data points')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Charges', fontsize=12)
plt.title('Scatter Plot of Age vs Insurance Charges', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

Using linear_model.LinearRegression() from sklearn, fit a model to your data, and make predictions on the data.

```python
# Initialize and fit a linear regression model
data_model = LinearRegression()
data_model.fit(x,y)

```

Plot another scatter plot with the best-fit line.

```python
# Plot the data and model
y_pred = data_model.predict(x)

# Scatter plot of actual data
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.6, edgecolor='k', label='Actual Data')

# Line plot of the regression model
x_sorted = np.sort(x, axis=0)
y_sorted = data_model.predict(x_sorted)
plt.plot(x_sorted, y_sorted, color='red', linewidth=2, label='Regression Line')

plt.xlabel('Age', fontsize=12)
plt.ylabel('Charges', fontsize=12)
plt.title('Age vs Insurance Charges with Regression Line', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()
```

We can see a relationship between age and insurance charges.
Meaning the older is the person the likely to pay more the insurance increases.

```python
# Scatter plot prediction with unknown data
unk_x = np.array([[65]]).reshape(-1,1) # Put the unknown value into 2d array

unk_y = data_model.predict(unk_x)

plt.figure(figsize=(10, 6))

# Scatter plot of actual data
plt.scatter(x, y, color='blue', alpha=0.6, edgecolor='k', label='Actual Data', s=40)

# Line plot of the regression model
x_sorted = np.sort(x, axis=0)
y_sorted = data_model.predict(x_sorted)
plt.plot(x_sorted, y_sorted, color='red', linewidth=2, label='Regression Line')

plt.scatter(unk_x, unk_y, color='green', edgecolor='k', s=60, label='Predicted Value (age 65)')

plt.xlabel('Age', fontsize=12)
plt.ylabel('Charges', fontsize=12)
plt.title('Age vs Insurance Charges with Regression Line', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()

print("The charge for a 65 year old person should be:", unk_y[0])
```

The prediction value output keeps on the same label that our regression line
