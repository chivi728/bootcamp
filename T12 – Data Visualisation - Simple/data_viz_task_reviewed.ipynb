[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
```

### 1. Boxplot

```python
# Boxplot for revs per mile
plt.figure(figsize=(10, 10)) # Size of the plot

# Another way using seaborn
#sns.boxplot(x='Rev.per.mile', y='Manufacturer', data=car_data)

# Group by manufactures and fetching the data
manufacturers = car_data['Manufacturer'].unique()
data = [car_data.loc[car_data['Manufacturer'] == m, 'Rev.per.mile'] for m in manufacturers]

# Create the boxplot and named the title, x and y labels
plt.boxplot(data, vert=False, tick_labels=manufacturers)
plt.gca().set_facecolor('skyblue')
plt.xlabel('Rev.per.mile')
plt.ylabel('Manufacturer')
plt.legend(['Revs per mile'])
plt.title('Box plot of Revolutions per Mile by Manufacturer')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:The manufactures between 3000 and 3500 RPM are:
** Toyota, Suzuki, Subaru, Pontiac, Geo, Ford, Dodge
**


### 2. Histogram

```python
# Histogram of MPG

# Size of the histogram
plt.figure(figsize=(10, 6))

#  Create histogram and named the title, x and y labels
plt.hist(car_data['MPG.city'], bins=10, color='skyblue', edgecolor='black', linewidth=1.2)
plt.hist(car_data['MPG.highway'], bins=10, color='green', edgecolor='black', linewidth=1.2)

plt.title('Histogram of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.legend(['City', 'Highway'])
plt.show()
plt.close()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:
According to the histogram drive on a highway is more efficient than on a city
**

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

# Size of the lineplot
plt.figure(figsize=(10, 6))

# Lineplot creation  and named the title, x and y labels
sns.lineplot(x='Wheelbase', y='Turn.circle',data=car_data)
plt.title('Wheelbase vs turning circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning circle')
plt.legend(['Wheelbase' ,'Turning circle'])
plt.show()
plt.close()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:
The relationships is as the wheelbase increases the turning circle becomes larger. This is both questions asked.
**


### 4. Barplot

```python
# Barplot for mean horsepower

# Size of the barplot
plt.figure(figsize=(24, 6))

# Calculation of the mean
mean_horsepower = car_data[['Horsepower', 'Manufacturer']].groupby('Manufacturer').mean()

# Barplot creation and named the title
sns.barplot(x='Manufacturer', y='Horsepower', data=car_data)
plt.title('Mean horsepower by manufacturer')
plt.legend(['Mean horsepower'])
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:
According to the diagram, yes. Bigger cars, more horsepower. But this is assumed a bit of knowledge of car brands.
**
